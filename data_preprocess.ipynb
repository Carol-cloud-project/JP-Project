```python
import pandas as pd
```

```python
# Initialize the stock list
stock_list = pd.read_csv('stock_list.csv')
df_X=pd.read_csv('df_X')
#
```

```python
stock_list=stock_list['stock']
```

```python
# Convert stock_list (RangeIndex) to a Python list
stock_list = stock_list.tolist()

# Initialize an empty list to store the new stock names
new_stock_list = []

# Track the current stock to be assigned
current_stock = stock_list[0]

# Loop over the rows of the dataframe
for i in range(len(df_X)):
    # If the current 'Day High' value is the same as the previous, keep the stock
    if i == 0 or df_X['Day High'][i] == df_X['Day High'][i-1]:
        new_stock_list.append(current_stock)
    else:
        # Update to the next stock in the list (cyclically)
        current_stock = stock_list[(stock_list.index(current_stock) + 1) % len(stock_list)]
        new_stock_list.append(current_stock)

# Assign the new stock list to the dataframe
df_X['new_stock'] = new_stock_list
```

```python
df_X.to_csv('df_X')
df_y=pd.read_csv('df_y')
df_y['new_stock']= new_stock_list
df_y.to_csv('df_y')
```
